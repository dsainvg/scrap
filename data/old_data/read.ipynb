```python
import json

# Read JSON file containing scraped links
with open('scraped_links.json', 'r', encoding='utf-8') as file:
    scraped_links = json.load(file)
```

```python
scraped_courses = scraped_links['course_pages']
```

```python
import pandas as pd

scraped_courses_df = pd.DataFrame(scraped_courses)

# Normalize URLs by converting all to https:// to avoid duplicates
scraped_courses_df['normalized_url'] = scraped_courses_df['url'].str.replace('http://', 'https://', regex=False)

# Get unique normalized URLs and sort them
unique_urls_df = pd.DataFrame(scraped_courses_df['normalized_url'].unique(), columns=['url']).sort_values(by='url')

# Save to CSV
unique_urls_df.to_csv('unique_course_urls.csv', index=False)

print(f"Total URLs in original data: {len(scraped_courses_df)}")
print(f"Unique URLs after normalization: {len(unique_urls_df)}")
```
